# K-mean clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_set= pd.read_csv("/Users/satyarth/Downloads/Clusterting_assignment.csv")
data_set
```

```python
data_set_p=data_set.values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
#WCSS is defined as the sum of the squared distance between each member of the cluster and its centroid.

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(data_set_p)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)# here optimum point is between 6 and 8
```

```python
kmeans = KMeans(n_clusters=2,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data_set_p)
y_kmeans
```

```python
my_clusters=pd.DataFrame({'clusters':range(1,11),'WSS':wcss})*.00001
my_clusters
```

```python
plt.scatter(data_set_p[y_kmeans==0,0],data_set_p[y_kmeans==0,1], s=100,c='red',label='cluster 1')
#here we are taking L as dependent variable and for y_kmeans we set zero first because it has value 0 in above result
#and zero in second place represents the index of x (first column of x) and 1 in second place in second term represents
#the second column of x
plt.scatter(data_set_p[y_kmeans==1,0],data_set_p[y_kmeans==1,1] ,s=100,c='blue',label='cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='orange')
```

# Spectral clustering

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

# Create synthetic data
data_set, labels = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
# here make blobs is used to make clusters

# Apply spectral clustering
spectral_clustering = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=42)
predicted_labels = spectral_clustering.fit_predict(data_set)

# Plot the results
plt.scatter(data_set[:, 0], data_set[:, 1], c=predicted_labels, s=50, cmap='viridis')
plt.show()
```

# Agglomerative clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Perform hierarchical clustering
Z = linkage(data_set, method='ward')
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# Define the number of clusters
k = 3

# Cut the dendrogram to obtain clusters
clusters = fcluster(Z, k, criterion='maxclust')

print("Assigned clusters:", clusters)
```

```python
# Plot the clusters
plt.scatter(data_set[:, 0], data_set[:, 1], c=clusters, cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering Results')
plt.show()
```

```python
Z = linkage(data_set, method='ward')

# Determine the number of clusters using the cophenetic distance
cophenetic_dist = np.max(data_set[:, 1])
max_d = cophenetic_dist * 0.7  # Adjust this threshold as needed
clusters = fcluster(Z, t=max_d, criterion='distance')
```

```python
silhouette_avg = silhouette_score(data_set, clusters)
print(f"Average silhouette score for hierarchical clusters: {silhouette_avg:.2f}")
```
